```python
from math import *
from sympy import Function, Symbol,sympify
import matplotlib.pyplot as plt
from scipy.integrate import quad


# Python code for simpson's 1 / 3 rule 
x = Symbol('x')
ff=open("lb001.txt","r")
p=ff.read()
xx=sympify(p, evaluate=False)
y=xx
print "input function ",y



def f(c):
    cc=float(y.subs({x:c}))
    return cc



def simpsons_( a, b, n ): 
    h = ( b-a )/n 
    c=[]
    x = list() 
    fx = list() 

    # Calcuting values of x and f(x) 
    i = 0
    while i<= n: 
        x.append(a + i * h) 
        fx.append(f(x[i])) 
        i += 1
    # Calculating result 
    res = 0
    i = 0
    tmp=[]
    while i<= n: 
        if i == 0 or i == n: 
            res+= fx[i] 
        elif i % 2 != 0: 
            res+= 4 * fx[i] 
        else: 
            res+= 2 * fx[i] 
        c.append(i)
        i+= 1
        tmp.append(res * (h / 3))
    res = res * (h / 3) 
    return res
    
    
    
def Trapezoidal(a, b, n):
    h = (b-a)/float(n)
    s = 0.5*(f(a) + f(b))
    for i in range(1,n,1):
        s = s + f(a + i*h)
    return h*s



a = 3.1 # Lower limit 
b = 3.9 # Upper limit 
n = 20 # Number of interval 

#
abres, err = quad(f, a,b)
print "standard result ",abres


cs=simpsons_(a,b, n)
cd=Trapezoidal(a,b, n)

print "I--(TRAPOEZO)  =   %f"%cd
print "I--(simpsons)  =    %f"%cs
```

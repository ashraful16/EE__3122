```python
import pandas as pd
import numpy as np
import scipy
import scipy.linalg as linalg 
import numpy as np
```

```python
xls=pd.ExcelFile("/media/ashraful/New Volume/labfiles/ex1d.xlsx")
print type(xls)
sheetx=xls.parse(0)
#print sheetx
var1=sheetx['aa']
var2=sheetx['bb']
var3=sheetx['cc']
var4=sheetx['dd']
#print ((var1), '\n ',var2, ' \n ',var3)
c=np.array([var1,var2,var3], dtype='float')
c= c.T
d=np.array(var4, dtype='float')
d=d.T
t=c
b=d
Ab = np.hstack([t, b.reshape(-1, 1)])
A=Ab
print A
#now we want to see how A has been factorized, P is the so called Permutation matrix
P, L, U = scipy.linalg.lu(A) 
print U
u=U
a3=1/(u[2][2]/u[2][3])
a21= u[1][3]-(u[1][2])*a3
a2=a21/u[1][1]
a11=u[0][3]-u[0][2]*a3-a2*u[0][1]
a1=a11/u[0][0]
print a1," \n ",a2," \n ",a3
```

```python
t=float(input("enter the value of t "))
y=a1*t*t+a2*t+a3

print y
```
